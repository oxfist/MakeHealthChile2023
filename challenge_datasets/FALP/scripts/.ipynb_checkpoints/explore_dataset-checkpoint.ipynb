# FALP cancer stage prediction challenge

## Dataset

This dataset contains a record of 19,110 cancer treatment services provided by the FALP to 11,023 patients between 2008 and 2022. Patient ID, demographic, diagnostic and treatment information is recorded for each case.

- Samples count: 19,110
- Patients: 11,023
- Distribution by gender: M (40.22498412410415%), F (59.77501587589585%)
- Deceased patients: 1,013 (9.189875714415313%)
- Cancer subcategories: 66

The size of the dataset is reduced when only patients with all reported TNM categories are considered.
- Patients with all TNM categories: 2,415
- Patients with all TNM categories and stage: 2,353
- Patients with all TNM categories, stage and tumor extension: 2,346

If only deceased patients are considered:
- Deceased patients with stage: 672
- Deceased patients with stage and tumor extension: 668
- Deceased patients with all TNM categories: 86
- Deceased patients with all TNM categories, stage and tumor extension: 85

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
from sklearn.cluster import KMeans
```

```python
path = os.path.join('..', 'RegistroTumoresFALP.csv')
data = pd.read_csv(path)
display(data)
```

```python
print(data.nunique())
```

```python
print("\n Categories: \n", data["CATEGORIA"].unique())
print("\n Subcategories: \n", data["SUBCATEGORIA"].unique())
print("\n Age: \n", data["EDAD"].unique())
print("\n Regions: \n", data["REGION"].unique())
print("\n Countys: \n", data["COMUNA"].unique())
print("\n Health insurance options: \n", data["PREVISION"].unique())
print("\n CT diagnostics: \n", data["CT"].unique())
print("\n CN diagnostics: \n", data["CN"].unique())
print("\n CM diagnostics: \n", data["CM"].unique())
print("\n PT diagnostics: \n", data["PT"].unique())
print("\n PN diagnostics: \n", data["PN"].unique())
print("\n PM diagnostics: \n", data["PM"].unique())
print("\n Stages: \n", data["ESTADIO"].unique())
print("\n Tumor extension options: \n", data["EXTENSION_DIAGNOSTICO"].unique())
print("\n Treatment services: \n", data["TTO_FALP_SUBCATEGORIA"].unique())
```

```python
start_years = [datetime.strptime(date, "%Y-%m-%d").year
               for date in data["FECHA_INICIO_TTO"]
               if date is not np.nan]
end_years = [datetime.strptime(date, "%Y-%m-%d").year
               for date in data["FECHA_FIN_TTO"]
               if date is not np.nan]
print(f"Cases between {min(start_years)} and {max(end_years)}")
```

```python
patients = data.drop_duplicates(subset='ID_CASO')
deceased = patients.loc[patients["ESTADO_VITAL"] == "FALLECIDO"]
print(patients.count())
masculine_count = patients.loc[patients["SEXO"] == 'M'].shape[0]
feminine_count = patients.loc[patients["SEXO"] == 'F'].shape[0]
print(f"\nPatient distribution by gender: M ({masculine_count*100/patients.shape[0]}%), F({feminine_count*100/patients.shape[0]}%)")
print(f"Deceased patients: {deceased.shape[0]} ({deceased.shape[0]*100/patients.shape[0]}%)")
display(deceased)
```

```python
subcat_distribution = {}
to_show = 20
for subcategory in patients["SUBCATEGORIA"].unique():
    subcat_distribution.update({
        subcategory: data.loc[data["SUBCATEGORIA"] == subcategory].shape[0]
    })
subcat_distribution = dict(sorted(subcat_distribution.items(), key=lambda x: x[1], reverse=True))
short_keys = [key[:50] for key in subcat_distribution.keys()]
_, ax = plt.subplots()
ax.barh(short_keys[:to_show], list(subcat_distribution.values())[:to_show])
for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.2,
             str(round((i.get_width()), 2)),
             fontsize = 6)
ax.set_xlabel('patients')
ax.set_title(f"{to_show} most frequent cancer subcategories in the dataset")
plt.show()
```

```python
region_distribution = {}
for region in patients["REGION"].unique():
    region_distribution.update({
        region: patients.loc[patients["REGION"] == region].shape[0]
    })
region_distribution = dict(sorted(region_distribution.items(), key=lambda x: x[1], reverse=True))
_, ax = plt.subplots()
ax.barh(list(region_distribution.keys()), list(region_distribution.values()))
for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.2,
             str(round((i.get_width()), 2)),
             fontsize = 6)
ax.set_xlabel('patients')
ax.set_title("Distribution of patients by region")
plt.show()
    
```

```python
with_tnm = patients.loc[
    ~patients["CT"].isna() &
    ~patients["CN"].isna() &
    ~patients["CM"].isna() &
    ~patients["PT"].isna() &
    ~patients["PN"].isna() &
    ~patients["PM"].isna()
]
with_tnm_and_stage = with_tnm.loc[
    ~with_tnm["ESTADIO"].isna()
]
with_all_diagnostic = with_tnm_and_stage.loc[ # any TNM, stage and extension
    ~with_tnm_and_stage["EXTENSION_DIAGNOSTICO"].isna()
]
deceased_with_all = with_all_diagnostic.loc[
    with_all_diagnostic["ESTADO_VITAL"] == "FALLECIDO"
]
deceased_with_stage = deceased.loc[
    ~deceased["ESTADIO"].isna()   
]
deceased_with_stage_extension = deceased.loc[
    ~deceased["ESTADIO"].isna() &
    ~deceased["EXTENSION_DIAGNOSTICO"].isna()
]
deceased_with_tnm = with_tnm.loc[
    with_tnm["ESTADO_VITAL"] == "FALLECIDO"
]
print(f"With all informed TNM: {with_tnm.shape[0]} patients")
print(f"With informed stage and all TNM: {with_tnm_and_stage.shape[0]} patients")
print(f"With informed extension, stage and all TNM: {with_all_diagnostic.shape[0]} patients")
print(f"Deceased with informed stage: {deceased_with_stage.shape[0]} patients")
print(f"Deceased with informed stage and extension: {deceased_with_stage_extension.shape[0]} patients")
print(f"Deceased with all TNM: {deceased_with_tnm.shape[0]} patients")
print(f"Deceased with all TNM, stage and extension: {deceased_with_all.shape[0]} patients")
```
