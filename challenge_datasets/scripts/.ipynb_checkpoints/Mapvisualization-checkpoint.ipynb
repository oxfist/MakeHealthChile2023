```python
import pandas as pd
import folium
import json
import numpy as np
```

```python
file = '../consolidado_covid19.csv'
df = pd.read_csv(file, index_col=0)
df.head()
```

```python
file = './locaciones_comuna.csv'
df_loc = pd.read_csv(file, index_col=0)
df_loc.head()
```

```python
df_region = df[df.Codigo_region == 10.0]
df_region.head()
```

```python
dates = pd.to_datetime(df_region.Fecha)
date_max = dates.max()

df_region.iloc[[-1]].Fecha
```

```python
reg = folium.Map(location=[-39.85,-72.8333], zoom_start=7,max_zoom=10,min_zoom=4,height=500,width="80%")
reg
```

```python
filtered_data = df_region[df_region.Fecha == '2022-08-12']

for i in range(0,len(df_region[df_region.Fecha == '2022-08-12'].Comuna)):
    
    comuna = filtered_data.Codigo_comuna.to_list()[i]
    comuna_name = filtered_data.Comuna.to_list()[i]
    loc_lat = df_loc[df_loc.Codigo_comuna == comuna].Latitud
    loc_lng = df_loc[df_loc.Codigo_comuna == comuna].Longitud
    
    total = filtered_data.Casos_incrementales.to_list()[i]

    folium.Circle(
        location=[loc_lat, loc_lng],
        
        tooltip = "<h5 style='text-align:center;font-weight: bold'>"+comuna_name+"</h5>"+
                    "<hr style='margin:10px;'>"+
                    "<ul style='color: #444;list-style-type:circle;align-item:left;padding-left:20px;padding-right:20px'>"+
        "<li>Confirmed: "+str(total)+"</li>"+
        "</ul>",
    
        radius=(int(np.log2(total+1)))*700,

         color='#ff6600',
        fill_color='#ff8533',
        fill=True).add_to(reg)
```
