### Health facilities dataset
- Replace mispelled values

```python
facilities_data = pd.read_excel('chile_sitios_salud.xlsx')
display(facilities_data)
```

```python
print(facilities_data.nunique())
for column in facilities_data.columns:
    print(column, ':')
    print(facilities_data[column].unique())
```

```python
# Fix mispelled values
facilities_data['PERTEN'] = facilities_data['PERTEN'].str.replace(r'^ Perteneciente$', 'Perteneciente', regex=True)
facilities_data['NOM_REG'] = facilities_data['NOM_REG'].str.replace(r'^Región De los Ríos$', 'Región De Los Ríos', regex=True)
facilities_data['CERTIFICA'] = facilities_data['CERTIFICA'].str.replace(r'^NO $', 'NO', regex=True)
facilities_data['DEPEN_A'] = facilities_data['DEPEN_A'].str.replace(r'^Municipal $', 'Municipal', regex=True)
facilities_data['NIVEL'] = facilities_data['NIVEL'].str.replace(r'^terciario$', 'Terciario', regex=True)
facilities_data['VIA'] = facilities_data['VIA'].str.replace(r'^calle$', 'Calle', regex=True)
facilities_data['VIA'] = facilities_data['VIA'].str.replace(r'^Calle $', 'Calle', regex=True)
facilities_data['SAPU'] = facilities_data['SAPU'].str.replace(r'^corto$', 'Corto', regex=True)
facilities_data['SAPU'] = facilities_data['SAPU'].str.replace(r'^no Aplica$', 'No Aplica', regex=True)
# Check duplicated
duplicated = facilities_data.duplicated()
print('duplicated rows: ', duplicated.sum())
# Save as csv
facilities_data.to_csv('chile_sitios_salud.csv', index=False)
```
